```python
import requests
import shutil
from tqdm import tqdm


years = [2020]
urls = set()

for year in years:
    for month in range(1,13):

        padded_month = str(month).rjust(2, "0")
        for i in range(10000):
            if (year == 2020 and month > 4):
                break
            padded_img = str(i).rjust(4, "0")

            urls.add(f"https://www.cherrycorals.com/wp-content/uploads/{year}/{padded_month}/CHE_{padded_img}.jpg")

print(urls)
            
            
for url in tqdm(urls):
    response = requests.get(url, stream=True)
    img_name = url.split("uploads/")[-1].replace("/", "_")

    # handle 404 and other issues
    if response.status_code == 404:
        continue
    if response.status_code == 429:
        print(f"{img_name} - issue with getting the image - status code {response.status_code}")
        continue
    
    # save result
    with open(f'./download/bc/{img_name}', 'wb') as out_file:
        shutil.copyfileobj(response.raw, out_file)
    del response
```
