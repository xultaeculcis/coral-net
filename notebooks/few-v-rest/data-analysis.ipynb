```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import skimage.io as io
import sys
from glob import glob
import os, os.path


%matplotlib inline

data_dir = "E:/Datasets/coral-classifier/few-v-rest/"


classes = os.listdir(data_dir)

dirs = [data_dir + "/" + c for c in classes ]
dirs
```
