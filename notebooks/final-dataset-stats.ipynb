```python
import os
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data_dir = "../datasets/labeled"
classes = sorted(os.listdir(data_dir))
classes
```

```python
class_dirs = [os.path.join(data_dir, c) for c in classes]
class_dirs
```

```python
class_counts = [len(items) for items in [os.listdir(cd) for cd in class_dirs]]
class_counts
```

```python
for c, count in  zip(classes, class_counts):
    print(f"{c}: {count}")
```

```python
fig = plt.figure(figsize=(12,7))
plt.bar(x=classes, height=class_counts)
plt.title('Image count per class')
plt.xlabel('Class')
plt.ylabel('Image instances')
plt.xticks(rotation=90)
plt.grid(color='#95a5a6', linestyle='--', linewidth=2, axis='y', alpha=0.7)
plt.show()
```

```python
# Sample images
from PIL import Image


samples = []
for c, class_dir in zip(classes, class_dirs):
    images = sorted(os.listdir(class_dir))
    image = images[-1]
    image = Image.open(os.path.join(class_dir,image))
    samples.append((c, image))
```

```python
def display_multiple_img(images, rows = 1, cols=1):
    figure, ax = plt.subplots(nrows=rows,ncols=cols, figsize=(10,20))
    for ind, t in enumerate(images):
        ax.ravel()[ind].imshow(t[1])
        ax.ravel()[ind].set_title(t[0])
        ax.ravel()[ind].set_axis_off()
    plt.tight_layout()
    plt.show()

display_multiple_img(samples, 13, 3)
```
